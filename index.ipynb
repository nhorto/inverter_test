## Join the weather and inverter data into the same dataframe
### and get the times only between 6 and 18

```python
# get hour for inverter dataframe
inv['Time'] = pd.to_datetime(inv['Time'])
inv['hour'] = inv['Time'].dt.hour

# get hour for weather dataframe
w['Time'] = pd.to_datetime(w['Time'])
w['hour'] = w['Time'].dt.hour

w = w.query('hour >= 6 and hour <= 18')
inv = inv.query('hour >= 6 and hour <= 18')
```

```python
# Sort by 'Time' (Required for merge_asof)
inv = inv.sort_values('Time')
w = w.sort_values('Time')

# Merge on closest timestamp
df = pd.merge_asof(inv, w, on='Time', direction='nearest')
df
```

```python
def fill_missing_solar_data(df, columns):
    """
    Fill short gaps in solar inverter data using linear interpolation
    
    Parameters:
    df: DataFrame with datetime index
    columns: List of columns to process
    """
    df_filled = df.copy()
    
    for column in columns:
        if column not in df.columns:
            print(f"Column '{column}' not found in DataFrame.")
            continue
            
        # Simple linear interpolation for gaps up to 4 points
        df_filled[column] = df_filled[column].interpolate(
            method='linear',
            limit=4  # Only fill gaps of 4 or fewer points
        )
    
    return df_filled
```

```python
def apply_swt_full_dataset(df, column="INVERTER 1 output_power_kW", wavelet="db4", level=2):
    """
    Apply SWT to the entire output power signal and store multiple levels of coefficients
    in the original DataFrame.

    Parameters:
        df (pd.DataFrame): DataFrame with a DateTime index.
        column (str): The column to process (e.g., output power).
        wavelet (str): Wavelet function to use for SWT.
        level (int): Number of decomposition levels.

    Returns:
        pd.DataFrame: DataFrame with new columns for A3, D1, D2, D3.
    """
    df = df.copy()
    
    signal = df[column].values

    # Ensure the signal is an even length by padding
    if len(signal) % 2 != 0:
        signal = np.pad(signal, (0, 1), mode="reflect")  # Reflect padding

    # Apply SWT (Full dataset)
    coeffs = pywt.swt(signal, wavelet, level=level)

    # Extract coefficients and map back to DataFrame
    df["A3"] = coeffs[2][0]  # Level 3 Approximation
    df["D1"] = coeffs[0][1]  # Level 1 Detail (Highest frequency)
    df["D2"] = coeffs[1][1]  # Level 2 Detail
    df["D3"] = coeffs[2][1]  # Level 3 Detail (Lowest frequency)

    return df

df_swt_full = apply_swt_full_dataset(tst)
```

```python
def combine_inverter_fields(df):
    metric_types = set(
        col.split("INVERTER 1 ")[-1] for col in df.columns if "INVERTER 1" in col
    )
    
    for metric in metric_types:
        columns_to_combine = [
            f"INVERTER {i} {metric}" for i in range(1, 4) 
            if f"INVERTER {i} {metric}" in df.columns
        ]
        
        # Check how many non-null values we have for each row
        non_null_counts = df[columns_to_combine].notna().sum(axis=1)
        
        if "power" in metric or "energy" in metric or "connections" in metric:
            df[f"total_{metric}"] = df[columns_to_combine].sum(axis=1)
            # Add count of contributing inverters
            df[f"contributing_inverters_{metric}"] = non_null_counts
        elif "temperature" in metric or "average" in metric:
            # Use pandas mean which automatically handles nulls
            df[f"average_{metric}"] = df[columns_to_combine].mean(axis=1)
            df[f"contributing_inverters_{metric}"] = non_null_counts
        else:
            df[f"total_{metric}"] = df[columns_to_combine].sum(axis=1)
            df[f"contributing_inverters_{metric}"] = non_null_counts
            
        df.drop(columns=columns_to_combine, inplace=True)

    return df

# Combine the fields in the dataset
combined_df = combine_inverter_fields(inv)
combined_df
```
